```python
import cv2 as cv
import numpy as np
from skimage import data, io, feature
from skimage.filters import threshold_local
from matplotlib import pyplot as plt
```

```python
img_frizzy = cv.imread('frizzy.png')
img_froomer = cv.imread('froomer.png')
```

```python
gray_frizzy = cv.cvtColor(img_frizzy, cv.COLOR_BGR2GRAY)
gray_froomer = cv.cvtColor(img_froomer, cv.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(20,20))
plt.subplot(1,2,1),plt.imshow(gray_frizzy,cmap = 'gray')
plt.title('Frizzy'), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2),plt.imshow(gray_froomer,cmap = 'gray')
plt.title('Froomer'), plt.xticks([]), plt.yticks([])
plt.show()
```

Canny edge detector: https://docs.opencv.org/3.1.0/dd/d1a/group__imgproc__feature.html#ga04723e007ed888ddf11d9ba04e2232de

```python
edges_frizzy = cv.Canny(gray_frizzy,30,35)
edges_froomer = cv.Canny(gray_froomer,30,35)
```

```python
#edges_frizzy = feature.canny(gray_frizzy)
#edges_froomer = feature.canny(gray_froomer)
```

```python
plt.figure(figsize=(20,20))
plt.subplot(1,2,1),plt.imshow(edges_frizzy,cmap = 'gray')
plt.title('Edges Frizzy'), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2),plt.imshow(edges_froomer,cmap = 'gray')
plt.title('Edges Froomer'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
common_edges = cv.bitwise_and(edges_frizzy,edges_froomer)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(common_edges,cmap = 'gray')
```
