```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv.imread('octagon.png',0)/255.0
```

```python
print(img)
```

```python
sobelx = cv.Sobel(img,cv.CV_64F,1,0,ksize=5)
sobely = cv.Sobel(img,cv.CV_64F,0,1,ksize=5)
```

```python
plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
gradient_magnitude = np.sqrt(sobelx**2 + sobely**2)
```

```python
scale_factor = np.max(gradient_magnitude)/255
gradient_magnitude = (gradient_magnitude/scale_factor).astype(np.uint8)
```

```python
plt.imshow(gradient_magnitude,cmap = 'gray')
plt.title('Edges of the image'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
abs_sobelx = np.absolute(sobelx)
abs_sobely = np.absolute(sobely)
absgraddir = np.arctan2(abs_sobely, abs_sobelx)
```

```python
plt.imshow(absgraddir,cmap = 'gray')
plt.title('Edges of the image'), plt.xticks([]), plt.yticks([])
plt.show()
```
