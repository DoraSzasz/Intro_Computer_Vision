```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

from scipy import signal
```

```python
img = cv2.imread('einstein.jpg')
```

```python
impulse = 2* signal.unit_impulse((3, 3), 'mid')
```

```python
print(impulse)
```

```python
box_filter =  np.ones((3,3),np.float32)/9
```

```python
print(box_filter)
```

```python
kernel = impulse - box_filter
```

```python
img_conv = cv2.filter2D(img,-1,kernel)
```

```python
plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(img_conv),plt.title('After linear filter')
plt.xticks([]), plt.yticks([])
plt.show()
```

What is the effect of applying this filter on the original image?
